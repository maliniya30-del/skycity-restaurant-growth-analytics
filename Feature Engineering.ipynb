```python
import pandas as pd
import numpy as np
df=pd.read_csv("data/SkyCity Auckland Restaurants & Bars.csv")
```

# Feature Engineering
## Core Business Metrics

### Total Revenue

```python
df['TotalRevenue']=(
    df['InStoreRevenue']
    + df['UberEatsRevenue']
    + df['DoorDashRevenue']
    + df['SelfDeliveryRevenue']
).round(0)
df['TotalRevenue'].describe()
```

### Total Profit

```python
df['TotalProfit']=(
     df['InStoreNetProfit']
    + df['UberEatsNetProfit']
    + df['DoorDashNetProfit']
    + df['SelfDeliveryNetProfit']
).round(0)
df['TotalProfit'].describe()
```

### Profit Margin

```python
df['ProfitMargin']=df['TotalProfit']/df['TotalRevenue']
df['ProfitMargin'].describe()
```

### Aggregator Dependence

```python
df['AggregatorDependence']= (df['UE_share']+df['DD_share'])
df['AggregatorDependence'].describe()
```

### Cost Burden

```python
df['CostBurden']=(df['COGSRate']+df['OPEXRate'])
df['CostBurden'].describe()
```

### Revenue Quality

```python
df['RevenueQuality']=df['AOV']*df['ProfitMargin']
df['RevenueQuality'].describe()
```

### Raw Scale Signal

```python
df['Raw_Scale_Signal']=(
    df['MonthlyOrders']*df['GrowthFactor']
)
df['Raw_Scale_Signal'].describe()
```

### Expansion Headroom

```python
df['Expansion_Headroom']=(
    df['DeliveryRadiusKM']* df['GrowthFactor']
)
df['Expansion_Headroom'].describe()
```

```python
feature_cols=[
    'TotalProfit','TotalRevenue','ProfitMargin','AggregatorDependence','CostBurden',
    'RevenueQuality',
    'Raw_Scale_Signal','Expansion_Headroom']
df[feature_cols]
```
